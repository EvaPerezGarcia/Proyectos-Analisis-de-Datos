## <span style="font-family: Palatino, 'Palatino Linotype', 'Book Antiqua', serif;"> **Introducción**

#### <span style="font-family: Palatino, 'Palatino Linotype', 'Book Antiqua', serif;"> **Índice de Desarrollo Humano (IDH)**

El Índice de Desarrollo Humano (IDH) es una medida compuesta desarrollada por el Programa de las Naciones Unidas para el Desarrollo (PNUD) que evalúa el nivel de desarrollo de un país a partir de tres dimensiones clave:

- Esperanza de vida al nacer (como indicador de salud).

- Promedio de años de escolarización y años esperados de escolarización (educación).

- Ingreso Nacional Bruto per cápita (nivel de vida o renta).

El IDH permite ir más allá de las métricas puramente económicas, ofreciendo una visión más integral del bienestar y las oportunidades reales de las personas para vivir una vida larga, saludable y con acceso al conocimiento y a los recursos necesarios.

#### <span style="font-family: Palatino, 'Palatino Linotype', 'Book Antiqua', serif;"> **¿Por qué es importante en el contexto europeo?**

Europa es una de las regiones del mundo con mayor desarrollo humano según los informes anuales del PNUD. Sin embargo, existen diferencias notables entre países del norte, sur, este y oeste del continente. Estas diferencias están influidas por múltiples factores como las políticas públicas, el sistema educativo, el acceso a servicios sanitarios y el nivel de ingresos.

Analizar el IDH en Europa permite identificar desigualdades regionales, evaluar la eficacia de las políticas sociales y comprender mejor el vínculo entre desarrollo y bienestar.

#### <span style="font-family: Palatino, 'Palatino Linotype', 'Book Antiqua', serif;"> **Relación entre el IDH y el bienestar social**

El bienestar social se refiere al nivel de satisfacción de las necesidades básicas de la población y al grado de calidad de vida en una sociedad. Aunque el IDH ya incluye componentes de salud, educación y renta, su análisis conjunto con otras variables sociales permite una comprensión más rica del contexto:

- Gasto público en sanidad y educación: puede mejorar directamente las dimensiones de salud y educación del IDH.

- Índice de Gini: mide la desigualdad de ingresos y puede influir en la percepción del bienestar.

- Niveles de pobreza o exclusión social: también están relacionados con el desarrollo humano.

Este estudio tiene como objetivo explorar cómo se distribuye el IDH entre los países europeos y qué relación guarda con diferentes indicadores de bienestar social, buscando patrones comunes o contrastes significativos.

## <span style="font-family: Palatino, 'Palatino Linotype', 'Book Antiqua', serif;"> **1. IDH (Índice de Desarrollo Humano)**

#### <span style="font-family: Palatino, 'Palatino Linotype', 'Book Antiqua', serif;"> **Top 10 países con mayor IDH en Europa**

```python
import pandas as pd
import matplotlib.pyplot as plt

url = "idheuropa.html"
tables = pd.read_html(url)
df = tables[0]

df.columns = ['Rango', 'País', 'IDH', 'Esperanza de vida', 'Educación', 'Ingreso', 'Categoria', 'Región', 'Año']  
df = df.drop(40, axis=0) 

def convertir_idh(idh_str):
    idh_str = str(idh_str).replace(',', '.') 
    if '.' in idh_str:
        return float(idh_str)
    else:
        return float('0.' + idh_str[1:])

df['IDH'] = df['IDH'].apply(convertir_idh)


top10 = df.sort_values(by='IDH', ascending=False).head(10) 
print("Top 10 países europeos por IDH:")
print(top10[['País', 'IDH']])
```

```python
plt.figure(figsize=(12.5, 6))
plt.bar(top10['País'][::-1], top10['IDH'][::-1], color='orange')

plt.ylim(0.94, 0.98) 

plt.xlabel("Índice de Desarrollo Humano (IDH)")
plt.title("Top 10 países europeos por IDH")
plt.grid(axis='x', linestyle='--', alpha=0.5)
plt.show()
```

<br>

Si quieres saber más acerca de un país concreto y consultar su Índice de Desarrollo Humano (IDH) en un año específico, puedes utilizar esta herramienta interactiva. Solo tienes que escribir el nombre del país tal como aparece en la tabla y el año disponible que deseas consultar. El programa te devolverá el valor exacto del IDH para ese país y ese año.

#### <span style="font-family: Palatino, 'Palatino Linotype', 'Book Antiqua', serif;"> **Consulta interactiva del IDH por país y año**

```python
df.columns = ['Rango', 'País', '2022', '2021', '2020', '2019', '2015', 'Puesto', 'Crecimiento']

pais = input("Introduce el nombre del país tal como aparece en la tabla: ")
año = input("Introduce uno de los años disponibles (2015, 2019, 2020, 2021, 2022):")

filtro = df['País'] == pais
df_pais = df[filtro]

if len(df_pais) > 0:
    valor_convertido = df_pais[año].values[0]
    idh = convertir_idh(valor_convertido)  
    print("El IDH de", pais, "en", año, "es:", idh)
else:
    print("Ese país no se ha encontrado.")
```

<br>

Con este programa hemos podido ver cómo ha cambiado el **Índice de Desarrollo Humano (IDH)** en los países de Europa según el año que el usuario quiera consultar. Esto es útil para comparar países y poder entender mejor sus diferencias en cuanto a desarrollo y bienestar.

Al ver el IDH en distintos países, se puede apreciar como los que tienen un IDH más bajo suelen ser también los que tienen más problemas económicos. Es decir, los países con menor IDH suelen ser los más pobres, con menos ingresos, menor acceso a educación y peor salud.

Esto demuestra que el desarrollo humano está muy relacionado con la economía, pero también con otros factores importantes como la educación y la sanidad.



#### <span style="font-family: Palatino, 'Palatino Linotype', 'Book Antiqua', serif;"> **Gráfico interactivo de la evolución del IDH por país**

Observar cómo ha evolucionado el IDH en los distintos países a lo largo del tiempo nos permite detectar mejoras, retrocesos o estancamientos en su desarrollo. Gracias al gráfico interactivo que hemos creado, es posible visualizar fácilmente esa evolución entre los años 2015 y 2022. El usuario solo tiene que introducir el nombre del país que quiere consultar, y el programa mostrará su recorrido en esos años de forma clara y visual.

## <span style="font-family: Palatino, 'Palatino Linotype', 'Book Antiqua', serif;"> **2. PIB (Producto Interior Bruto)**

#### <span style="font-family: Palatino, 'Palatino Linotype', 'Book Antiqua', serif;"> **Relación entre el PIB y el IDH**

El **Producto Interior Bruto (PIB) per cápita** es uno de los indicadores económicos más utilizados para medir la riqueza de un país. Aunque por sí solo no refleja todos los aspectos del desarrollo humano, tiene una relación directa con el nivel de vida de la población, uno de los tres factores clave que se tienen en cuenta para calcular el IDH.

Por ello es importante, analizar el PIB per cápita de los países europeos con el objetivo de compararlo con los datos del IDH y ver cómo la economía influye en el bienestar general. Para facilitar este análisis, se ha incluido una visualización interactiva en forma de mapa de calor, en la que el usuario puede seleccionar un año concreto y observar cómo varía el PIB entre países.

```python
df_pib = pd.read_csv("estat_tec00001_filtered_en.csv")
df_pib = df_pib.drop(columns=["DATAFLOW", "LAST UPDATE", "freq", "na_item", "unit", "OBS_FLAG", "CONF_STATUS"])

print(df_pib.columns)
print(df_pib.head())
```

#### <span style="font-family: Palatino, 'Palatino Linotype', 'Book Antiqua', serif;"> **Mapa de calor del PIB per cápita interactivo**

Ahora tienes la oportunidad de observar de forma visual cómo se distribuye el PIB per cápita entre los países europeos en un año concreto. A través de este mapa de calor, podrás identificar fácilmente qué países tienen un mayor nivel económico y cuáles presentan valores más bajos. Solo hace falta escribir el año que se desea consultar, y el gráfico mostrará las diferencias.

```python
año = input("Introduce el año que quieres visualizar: ")

df_filtrado = df_pib[df_pib['TIME_PERIOD'] == int(año)]
df_filtrado = df_filtrado.sort_values(by='OBS_VALUE', ascending=False)

matriz = []
for fila in df_filtrado['OBS_VALUE']:
    matriz.append([fila])

plt.figure(figsize=(5, 10))
plt.imshow(matriz, cmap='Oranges', aspect='auto')
plt.colorbar(label="PIB per cápita (€)")
plt.title("Mapa de calor del PIB per cápita en el año " + año)
plt.yticks(range(len(df_filtrado)), df_filtrado['geo'])
plt.xticks([0], [año])
plt.show()
```

Gracias a este gráfico, se puede apreciar con facilidad la desigualdad económica entre los países europeos. Algunos presentan valores muy altos de PIB per cápita, mientras que otros siguen muy por debajo de la media. Esto refleja diferencias importantes en el nivel de vida y en el desarrollo económico de cada país.

#### <span style="font-family: Palatino, 'Palatino Linotype', 'Book Antiqua', serif;"> **Gráfico interactivo de la evolución del PIB per cápita entre 2 países**

Ahora tienes la oportunidad de comparar la evolución del PIB per cápita entre dos países europeos. Solo tienes que escribir los nombres de los países que te interesen, y el gráfico mostrará cómo ha cambiado su nivel económico a lo largo de los años. Esta herramienta permite ver de forma visual si han seguido trayectorias parecidas o si ha habido diferencias importantes entre ellos.

```python
pais1 = input("Introduce el primer país (debe estar escrito en inglés):")
pais2 = input("Introduce el segundo país (debe estar escrito en inglés):")

ls_agnos = []
for fila in df_pib['TIME_PERIOD']:
    if fila not in ls_agnos:
        ls_agnos.append(fila)
ls_agnos.sort()

ls_pais1 = []
ls_pais2 = []

for agno in ls_agnos:
    #PIB del país 1
    dato1 = df_pib[(df_pib['geo'] == pais1) & (df_pib['TIME_PERIOD'] == agno)]
    if len(dato1) > 0:
        ls_pais1.append(dato1['OBS_VALUE'].iloc[0])
    else:
        ls_pais1.append(None)

    #PIB del país 2
    dato2 = df_pib[(df_pib['geo'] == pais2) & (df_pib['TIME_PERIOD'] == agno)]
    if len(dato2) > 0:
        ls_pais2.append(dato2['OBS_VALUE'].iloc[0])
    else:
        ls_pais2.append(None)

fig, ax = plt.subplots()

ax.plot(ls_pais1, label=pais1)
ax.plot(ls_pais2, label=pais2)
ax.legend()

ax.set(xlabel='Años', ylabel='PIB per cápita (€)', title='Evolución del PIB per cápita')
ax.set_xticks(range(len(ls_agnos)))
ax.set_xticklabels(ls_agnos, rotation="vertical")

plt.show()
```

#### <span style="font-family: Palatino, 'Palatino Linotype', 'Book Antiqua', serif;"> **Consulta de la media del PIB per cápita en Europa**

Ahora puedes consultar la media del PIB per cápita de Europa en el año que elijas. Solo tienes que escribir el año que te interesa, y el programa calculará la media de todos los países europeos en ese periodo. Es una forma rápida de conocer el nivel económico general del continente sin necesidad de gráficos.

```python
año = input("Introduce el año que quieres consultar: ")

df_año = df_pib[df_pib['TIME_PERIOD'] == int(año)]

suma = 0
contador = 0

for valor in df_año['OBS_VALUE']:
    suma += valor
    contador += 1

#Media de los países
if contador > 0:
    media = suma / contador
    print("La media del PIB per cápita en Europa en el año", año, "fue de:", round(media, 2), "euros.") 
else:
    print("No hay datos disponibles para ese año.")
```

#### <span style="font-family: Palatino, 'Palatino Linotype', 'Book Antiqua', serif;"> **Comparación del PIB de un país con la media europea**

Después de calcular la media del PIB per cápita en Europa, damos un paso más comparando ese valor con el PIB de un país concreto. Esta consulta permite saber si ese país está por encima o por debajo de la media europea en el mismo año, y en qué medida.

```python
pais = input("Introduce el país (debe estar escrito en inglés): ")
df_pais = df_año[df_año['geo'] == pais]

if len(df_pais) > 0:
    pib = df_pais['OBS_VALUE'].iloc[0] 

    if pib > media:
        print(pais, "está por ENCIMA de la media europea en", año, "con", round(pib - media, 2), "euros más.")
    elif pib < media:
        print(pais, "está por DEBAJO de la media europea en", año, "con", round(media - pib, 2), "euros menos.")
    else:
        print(pais, "tiene un PIB igual a la media europea en", año)
else:
    print("No se han encontrado datos para ese país en ese año.")
```

## <span style="font-family: Palatino, 'Palatino Linotype', 'Book Antiqua', serif;"> **3. Índice de natalidad**

Pero cuando miramos el índice de natalidad, nos damos cuenta de que no siempre va de la mano. Hay países con un PIB muy alto que, sin embargo, tienen tasas de natalidad bastante bajas. #### <span style="font-family: Palatino, 'Palatino Linotype', 'Book Antiqua', serif;"> **Unión de 2 Datasets: PIB per cápita y las tasa de natalidad**

```python
df_pib.rename(columns={'geo': 'País'}, inplace=True)
df_demo.rename(columns={'geo': 'País'}, inplace=True)

df_unido = pd.merge(df_pib, df_demo, on=['País', 'TIME_PERIOD'], how='outer')
print(df_unido)
```

#### <span style="font-family: Palatino, 'Palatino Linotype', 'Book Antiqua', serif;"> **Relación entre natalidad y PIB per cápita a lo largo del tiempo**

En este apartado hemos querido observar si existe una relación entre el crecimiento económico y la evolución del índice de natalidad en un mismo país. Para ello, hemos unido en una sola tabla los datos de PIB per cápita y natalidad, y los hemos representado gráficamente. Cada punto del gráfico muestra el PIB y la natalidad de un año concreto, lo que nos permite ver cómo se han movido estos dos indicadores a lo largo del tiempo. De este modo, se puede intuir si a medida que un país mejora económicamente, la natalidad sube, baja o se mantiene estable.

```python
pais = input("Introduce el nombre del país (en inglés): ")
año_inicio = int(input("Introduce el primer año: "))
año_final = int(input("Introduce el último año: "))

df_pais = df_unido[(df_unido['País'] == pais) & 
                   (df_unido['TIME_PERIOD'] >= año_inicio) & 
                   (df_unido['TIME_PERIOD'] <= año_final)]

x_natalidad = []
y_pib = []
etiquetas = []

for i in range(len(df_pais)):
    natalidad = df_pais['OBS_VALUE_y'].iloc[i]
    pib = df_pais['OBS_VALUE_x'].iloc[i]
    año = df_pais['TIME_PERIOD'].iloc[i]
    
    if natalidad == natalidad and pib == pib:  
        x_natalidad.append(natalidad)
        y_pib.append(pib)
        etiquetas.append(año)

if len(x_natalidad) > 0:
    plt.figure(figsize=(8, 6))
    plt.scatter(x_natalidad, y_pib, color='orange', s=300, alpha=0.7)

    for i in range(len(etiquetas)):
        plt.text(x_natalidad[i], y_pib[i], str(etiquetas[i]), 
                 color='black', ha='center', va='center', fontsize=9)

    plt.xlabel("Índice de natalidad")
    plt.ylabel("PIB per cápita (€)")
    plt.title(f"Natalidad vs PIB en {pais} ({año_inicio} - {año_final})")
    plt.grid(True)
    plt.show()
else:
    print("No hay datos suficientes para ese país y ese rango de años.")
```

#### <span style="font-family: Palatino, 'Palatino Linotype', 'Book Antiqua', serif;"> **¿Un país está por encima o por debajo de la media europea de natalidad?**

```python
año = int(input("Introduce el año: "))
pais = input("Introduce el país (tiene que estar en inglés): ")

df_año = df_unido[df_unido['TIME_PERIOD'] == año]

suma = 0
contador = 0

for valor in df_año['OBS_VALUE_y']:
    if valor == valor:
        suma += valor
        contador += 1

media = suma / contador if contador > 0 else 0

df_pais = df_año[df_año['País'] == pais]
if len(df_pais) > 0:
    for valor in df_pais['OBS_VALUE_y']:
        break  
    
    print(f"Índice de natalidad de {pais} en {año}: {valor}")
    print(f"Media europea: {round(media, 2)}")
    
    if valor > media:
        print(f"{pais} está por encima de la media.")
    elif valor < media:
        print(f"{pais} está por debajo de la media.")
    else:
        print(f"{pais} está justo en la media.")
```

#### <span style="font-family: Palatino, 'Palatino Linotype', 'Book Antiqua', serif;"> **¿Cuál ha sido el año con mayor natalidad en un país?**

```python
pais = input("Introduce el nombre del país (en inglés): ")
df_pais = df_unido[df_unido['País'] == pais]

mayor_valor = 0
año_maximo = None

for i in range(len(df_pais)):
    natalidad = df_pais['OBS_VALUE_y'].iloc[i]     
    año = df_pais['TIME_PERIOD'].iloc[i]            

    if natalidad == natalidad:
        if natalidad > mayor_valor:
            mayor_valor = natalidad
            año_maximo = año

if año_maximo is not None:
    print(f"El año con mayor natalidad en {pais} fue {año_maximo} con un valor de {mayor_valor}.")
else:
    print("No hay datos disponibles de natalidad para ese país.")
```

Con estos dos últimos ejercicios hemos podido investigar un poco más cómo está un país respecto al resto de Europa en cuanto a natalidad. Saber si está por encima o por debajo de la media ayuda a ver si sigue la tendencia general o si destaca por algo. Además, ver en qué año tuvo su natalidad más alta nos da una pista de cuándo hubo más nacimientos, que puede estar relacionado con cambios sociales, ayudas, o simplemente una etapa de crecimiento.
